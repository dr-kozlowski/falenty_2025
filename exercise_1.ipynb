```python
T = 2e1 # ms
dt = 0.1 # ms
N = int(T / dt) # number of steps

c0 = 0 # initial concentration in the cell [num of ions / length**3]
Conc = np.zeros(N) # initialize an array to store the results
Conc[0] = c0 # put the initial value
j_in = lambda j, t: j * np.sin(t) if j * np.sin(t) > 0 else 0
j_out = lambda t, Conc: k * Conc
k = 0.5

# C_i = lambda t, C_i0: C_i0 * np.sin(t) if C_i0 * np.sin(t) > 0 else 0


C_i0 = 1
Conc_i = np.zeros(N)
Conc_i[0] = C_i0

ji_in = lambda m, Conc, t: m * Conc
ji_out = lambda k, Conc_i, t: k * Conc_i
m = 0.5

j_ins = []
Ts = []
```

```python
for i in range(0, N):
    t = i * dt
    Conc[i] = Conc[i-1] + dt * (j_in(j=0.6, t=t) - j_out(t, Conc[i-1]) + Conc_i[i-1])
    Conc_i[i] = ji_in(m=m, Conc=Conc[i-1], t=t) - ji_out(k, Conc_i[i-1], t)
```

```python
ts = np.linspace(0, T, N)
pts = 150
plt.plot(ts[:pts], Conc[:pts])
plt.xlabel("Time")
plt.ylabel("Concentration")
plt.grid(True)
plt.show()
```
